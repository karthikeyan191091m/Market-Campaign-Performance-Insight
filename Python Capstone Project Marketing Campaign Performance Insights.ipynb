# **3) Exploratory Data Analysis (EDA) and Visualization**

Campaign Performance:



*   Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI



```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df["Acquisition_Cost"],df["ROI"])
plt.title("Acquisition Cost vs ROI")
plt.xlabel("Acquisition Cost")
plt.ylabel("ROI")
plt.show()
```



*   Create a bar chat to visualize the average Conversion_Rate for different Channel_Used, categorized by Campaign_Type.



```python
avg_conv = df.groupby("Channel_Used")["Conversion_Rate"].mean()

bars = plt.bar(avg_conv.index, avg_conv.values)
plt.title("Average Conversion Rate by Channel")
plt.xlabel("Channel")
plt.ylabel("Conversion Rate")
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval,2),
             ha="center", va="bottom")

plt.xticks(rotation=45)
plt.show()
```



*   Visualize the distribution of Engagement_Score across different Campaign_Type using a box plot



```python
df.boxplot(column="Engagement_Score", by="Campaign_Type")
plt.title("Engagement Score by Campaign Type")
plt.suptitle("")

means = df.groupby("Campaign_Type")["Engagement_Score"].mean()
for i, val in enumerate(means):
    plt.text(i+1, val, round(val,2), ha="center", va="bottom", color="blue")

plt.show()
```



*  Analyze the average ROI by Company using a bar chart to compare the profitability of campaigns conducted by different companies



```python
avg_roi = df.groupby("Company")["ROI"].mean().sort_values(ascending=False)
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b"]
ax = avg_roi.plot(kind="bar", title="Average ROI by Company", color=colors[:len(avg_roi)])
for container in ax.containers:
    ax.bar_label(container, padding=3, fontsize=10, color='black', fontweight='bold')

plt.show()
```



*   Examine the correlation between Engagment_Score and Conversion_Rate using a heatmap



```python
import numpy as np
```

```python
corr = df[["Engagement_Score", "Conversion_Rate"]].corr()
plt.imshow(corr, cmap="Blues", interpolation="nearest")
plt.xticks([0,1], ["Engagement_Score", "Conversion_Rate"])
plt.yticks([0,1], ["Engagement_Score", "Conversion_Rate"])
plt.colorbar()
plt.title("Correlation Heatmap")
plt.show()
```

# **Customer Segmentation**



*   Create a count plot to visualize the distribution of Target_Audience



```python
df["Target_Audience"].value_counts().plot(kind="bar", title="Target Audience Distribution")
plt.show()
```



*   Identify which Customer_Segment has the highest Conversion_Rate for each Language using a bar chat



```python
result = df.groupby(["Language", "Customer_Segment"])["Conversion_Rate"].mean()
result_unstacked = result.unstack()
result_unstacked.plot(kind='bar')
plt.xlabel("Language")
plt.ylabel("Average Conversion Rate")
plt.title("Average Conversion Rate by Language and Customer Segment")
plt.legend(title="Customer Segment")
plt.show()
```



*   Visualize the distribution of Acquisition_Cost across each Customer_Segment, categorized by Channel_Used, using a box plot.



```python
import seaborn as sns
```

```python
custom_palette = {
    'Google Ads': '#1f77b4','Website': '#ff7f0e','Email': '#2ca02c','YouTube': '#d62728','Facebook': '#9467bd','Instagram': '#8c564b'}
plt.figure(figsize=(8, 4), facecolor='white')
ax = sns.boxplot(
    x='Customer_Segment',y='Acquisition_Cost',hue='Channel_Used',data=df,palette=custom_palette)
plt.xlabel('Customer Segment')
plt.ylabel('Acquisition Cost')
plt.xticks(rotation=45)
plt.title('Distribution of Acquisition Cost by Customer Segment and Channel Used', color='black')
plt.legend(title='Channel Used', loc="upper left", bbox_to_anchor=(1, 1))

plt.show()
```



*   Analyze average Conversion_Rate by Language using a bar chart to compare the effective of compaigns conducted in different languages.



```python
avg_lang = df.groupby("Language")["Conversion_Rate"].mean()
avg_lang.plot(kind="bar", title="Average Conversion Rate by Language")
plt.show()
```

# **Channel Effectiveness:**



*   Compare the Engagement_Score for different Channels_Used, segmented by Campaign_Type, using a bar chart.



```python
avg_eng = df.groupby("Channel_Used")["Engagement_Score"].mean()
avg_eng.plot(kind="bar", title="Average Engagement Score by Channel")
plt.show()
```



*   Show the distribution of total ROI across different Channels_Used using a pie chart.



```python
roi_dist = df.groupby("Channel_Used")["ROI"].sum()
plt.pie(roi_dist, labels=roi_dist.index, autopct="%1.1f%%")
plt.title("ROI Distribution by Channel")
plt.show()
```

# **Time Based Analysis:**



*   Plot the distribution of duration using a histogram



```python
plt.hist(df["Duration"], bins=15)
plt.title("Distribution of Campaign Duration")
plt.xlabel("Duration (days)")
plt.ylabel("Frequency")
plt.show()
```



*   Analyze how the overall Conversion_Rate has changed over Date for each Company using a line chart



```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
trend = df.groupby("Date")["Conversion_Rate"].mean()
trend.plot(title="Conversion Rate Over Time")
plt.show()
```



*   Examine the trend of Engagement_Score over Data with a line chart



```python
eng_trend = df.groupby("Date")["Engagement_Score"].mean()
eng_trend.plot(title="Engagement Score Over Time Trend")
plt.show()
```

# **Geographic Analysis:**



*   Determine which location has the highest Acquisition_Cost using a bar char



```python
avg_loc = df.groupby("Location")["Acquisition_Cost"].mean().sort_values(ascending=False)
avg_loc.plot(kind="bar", title="Average Acquisition Cost by Location")
plt.show()
```



*   Visualize the Conversion_Rate by different Location, categorized by Target_Audience, using a bar chart.



```python
pivot = df.pivot_table(index="Location", columns="Target_Audience", values="Conversion_Rate", aggfunc="mean")
pivot.plot(kind="bar", title="Conversion Rate by Location and Target Audience")
plt.show()
```



*   Illustrate the proportion of ROI by Location using a pie chart



```python
roi_loc = df.groupby("Location")["ROI"].sum()
plt.pie(roi_loc, labels=roi_loc.index, autopct="%1.1f%%")
plt.title("ROI Distribution by Location")
plt.show()
```
